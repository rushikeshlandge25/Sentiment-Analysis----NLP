```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

##Removing special characters
import re
df['reviewText']=df['reviewText'].apply(lambda x:re.sub(r'[^a-z A-Z 0-9-]', ' ', x))

#removing stopwords
df['reviewText'] = df['reviewText'].apply(lambda x: ' '.join([y for y in x.split() if y not in (stopwords.words('english'))]))
##remove url
df['reviewText']=df['reviewText'].apply(lambda x:re.sub(r'(http|https|ssh|ftp)://([\W_-]+(?:(?:\.[\w_-]+))([w.,@?^=%&/~+#-]))?', '', str(x)))
#remove tags
from bs4 import BeautifulSoup
df['reviewText']=df['reviewText'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())
df['reviewText']=df['reviewText'].apply(lambda x:" ".join(x.split()))
```

```python
df.head()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer_instance = WordNetLemmatizer()

def lemmatize_words(text):
  lemmatized_words = []
  for word in text.split():
    lemmatized_words.append(lemmatizer_instance.lemmatize(word))
  return " ".join(lemmatized_words)

df['reviewText']=df['reviewText'].apply(lambda x:lemmatize_words(x))
df.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],test_size=0.2,random_state=42)
```

### Feature Extraction: TF-IDF Vectorization

Before training a machine learning model, the text data needs to be converted into numerical features. TF-IDF (Term Frequency-Inverse Document Frequency) is a widely used technique for this, which reflects how important a word is to a document in a collection or corpus.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000) # You can adjust max_features based on your data and computational resources

# Fit and transform the training data
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

# Transform the test data using the *fitted* vectorizer
X_test_tfidf = tfidf_vectorizer.transform(X_test)

print(f"Shape of X_train_tfidf: {X_train_tfidf.shape}")
print(f"Shape of X_test_tfidf: {X_test_tfidf.shape}")
```

### Logistic Regression Algorithm

Logistic Regression is a linear model for binary classification problems. Despite its name, it is a classification algorithm rather than a regression algorithm. It uses a logistic function to model the probability of a certain class or event. In the context of text classification, it's often used as a strong baseline due to its simplicity, interpretability, and efficiency. When combined with TF-IDF features, it can perform quite well on many text classification tasks.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize and train the Logistic Regression model
# Set solver to 'liblinear' for smaller datasets or 'saga' for larger ones with L1 regularization
# 'C' is the inverse of regularization strength; smaller values specify stronger regularization.
lr_model = LogisticRegression(solver='liblinear', random_state=42, C=1.0)
lr_model.fit(X_train_tfidf, y_train)

# Predict on the test set
y_pred_lr = lr_model.predict(X_test_tfidf)

# Evaluate the Logistic Regression model
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f"Logistic Regression Accuracy: {accuracy_lr:.4f}")
print("\nLogistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))
print("\nLogistic Regression Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))
```
